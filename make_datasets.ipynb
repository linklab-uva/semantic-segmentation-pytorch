```python
import pandas as pd
```

```python
t = pd.read_csv('accuracies_iou_val2.txt')
```

```python
def to_odgt(tbl, file=None):
    out = pd.DataFrame({'fpath_img': tbl.File})
    out['fpath_segm'] = tbl.File.replace(r'images/(.*)\.jpg', r'labels/\1_train_id.png', regex=True)
    out['width'] = 1280
    out['height'] = 720
    if file is None:
        return out
    out.to_json(file, orient='records', lines=True)
```

```python
bad2 = t.sort_values('Acc')[:100]
to_odgt(bad2, '../bdd100k/seg/bad_val.odgt')
good2 = t.sort_values('Acc')[-100:]
to_odgt(good2, '../bdd100k/seg/good_val.odgt')
```

```python
bad3 = t[t.Car_union > 1000].nsmallest(100, 'Car_iou')
good3 = t[t.Car_union > 1000].nlargest(100, 'Car_iou')
to_odgt(bad3, '../bdd100k/seg/bad_car_val.odgt')
to_odgt(good3, '../bdd100k/seg/good_car_val.odgt')
```

```python
combined_u = t[['Car_union', 'Truck_union', 'Bus_union', 'Motorcycle_union', 'Bicycle_union', 'Person_union', 'Rider_union']].rename(columns=lambda s: s.split('_')[0])
combined_iou = t[['Car_iou', 'Truck_iou', 'Bus_iou', 'Motorcycle_iou', 'Bicycle_iou', 'Person_iou', 'Rider_iou']].rename(columns=lambda s: s.split('_')[0])
combined_i = combined_u * combined_iou
tot_u = combined_u.sum(1)
tot_iou = combined_i.sum(1) / tot_u
bad4 = t.iloc[tot_iou[tot_u > 1000].nsmallest(100).index]
to_odgt(bad4, '../bdd100k/seg/bad_combined_val.odgt')
bad4
good4 = t.iloc[tot_iou[tot_u > 1000].nlargest(100).index]
to_odgt(good4, '../bdd100k/seg/good_combined_val.odgt')
good4
```
